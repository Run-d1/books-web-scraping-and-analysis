## - Automate the Book Data Extraction Process

```python
def extract_book_data(book_soup):
    book_id = book_soup.find('table').find_all('td')[0].text
    category = book_soup.find('ul', class_='breadcrumb').find_all('li')[2].a.text
    quantity = book_soup.find('table').find_all('td')[5].text.split()[2].lstrip('(')
    return book_id, category, quantity
```

# Start Scraping (50) pages

```python
titles, prices, ratings, availabilities = [], [], [], []
```

```python
ids, categories, quantities = [], [], []
```

```python
for n in range(1, 51):
    page = requests.get(url.format(n))
    
    soup = BeautifulSoup(page.text, 'html')
    books = soup.find_all('article')
    
    for book in books:
        title = book.h3.a.text
        titles.append(title)
        print(title)
        
        price = book.find('div', class_='product_price').p.text[2:]
        prices.append(float(price))
        # print(price)
    
        rating = book.p['class'][1]    
        ratings.append(rating)
        # print(rating)
    
        availability = book.find('div', class_="product_price").find('p', class_='instock availability').text.strip()
        availabilities.append(availability)
        # print(availability)

        base_url = 'https://books.toscrape.com/catalogue/'
        
        book_url = book.h3.a['href']
        # print(book_url)  # >>> a-light-in-the-attic_1000/index.html
        
        book_page = requests.get(base_url + book_url)
        # print(book_page)  # >>> <Response [200]>
        
        book_soup = BeautifulSoup(book_page.text, 'html')
        # print(book_soup.prettify())

        book_id, category, quantity = extract_book_data(book_soup)
        print(quantity)
        ids.append(book_id)
        categories.append(category)
        quantities.append(quantity)

print(len(titles))
print(len(quantities))
```

# Crete the DataFrame

```python
df = pd.DataFrame({
    'ID': ids,
    'Title': titles,
    'Category': categories,
    'Price [£]': prices,
    'Rating': ratings,
    'Availability': availabilities,
    'Quantity': quantities
})

df
```

# Save DataFrame to CSV file

```python
df.to_csv('books-data.csv', index=False)
```
